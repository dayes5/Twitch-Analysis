## Background & Prior Work

The pandemic suddenly and drastically changed the way we interact with each other. People started to rely heavily on media technology more than ever to remain connected. This switch to online platforms as the primary source of staying in touch impacted the way we think of relationships. Considering how media technology became the main source of communication with loved ones during the pandemic, our relationships with media figures may come to mimic the way we process our relationships with real-life friends and family, with both occurring over screens. Furthermore, “the data suggest that the more time we spend communicating with our real-life friends via screens, the stronger we begin to perceive our parasocial relationships with media figures" (Psypost). 

Our heavy reliance on screen communication, as well as parasocial relationships’ strong resemblance of our personal relationships, have potentially blurred cognitive distinctions between loved ones in real life and media figures during the pandemic. 

In particular, the global pandemic during 2020 prompted a huge wave of people with other interests to take up live-streaming and/or watch streamers. The senior vice president of content at Twitch said “People are looking for a way to connect and when someone finds a community on Twitch, you become a regular viewer”(WSJ article). This leads us to believe that there is an increase in parasocial relationships between viewers and streamers on Twitch during COVID (a time of high reliance on screen communication). In our project we considered the commitment and start of a parasocial, one sided, relationship as a “follow”. When twitch users follow a streamer, they are creating a stronger commitment.

# Data Cleaning

Describe your data cleaning steps here.

```python
twitch2019And2020Data[0]
```

The data in the dataframes currently contains unnamed columns and columns which will not be used in the data analysis section. Therefore, we will be dropping the unnamed columns and the Partnered, Channel, Mature, Language, and Peek viewers columns for each dataframe. 

```python
# drop columns that are non numerical and not necessary for our project

for i in range(len(twitch2019And2020Data)):
    
    twitch2019And2020Data[i] = twitch2019And2020Data[i].loc[:, ~twitch2019And2020Data[i].columns.str.contains('^Unnamed')]
    twitch2019And2020Data[i] = twitch2019And2020Data[i].drop(columns=['Partnered', 'Channel', 'Mature', 'Language', 'Peak viewers', 'Average viewers'])

twitch2019And2020Data[0]
```

The dataframes now only contain all the necessary columns needed for analysis. Still, the followers gained column does not represent standardize data since our data consists of huge streamers and smaller ones. Thus, we need to standardize the data by obtaining the percentage of followers gained for each streamer. 

```python
# Calculating the percent of followers gained and renaming the column
for i in range(len(twitch2019And2020Data)):
    twitch2019And2020Data[i]['Followers gained'] = twitch2019And2020Data[i]['Followers gained']/twitch2019And2020Data[i]['Followers']*100
    twitch2019And2020Data[i] = twitch2019And2020Data[i].rename(columns={"Followers gained": "Followers gained percent"})

twitch2019And2020Data[0]
```

Now that we have standardize the data, we are able to calculate the mean for each dataset.

```python
# calculate mean of each dataset and input into new data frame
meanOfTwitchData= []
for i in range(len(twitch2019And2020Data)):
    meanOfTwitchData.append(twitch2019And2020Data[i].mean())      
meanOfTwitchData[0]
```

Since all of the dataframes in meanOfTwitchData have the same columns names and shape size, we are able to merge them into one dataframe. 

```python
meanOf2019Data = []
meanOf2020Data = []
for i in range(len(meanOfTwitchData)):
    if (i < 12):
        meanOf2019Data.append(meanOfTwitchData[i])
    else:
        meanOf2020Data.append(meanOfTwitchData[i])
    
```

```python
df = pd.DataFrame(data=meanOfTwitchData) 
df2019 = pd.DataFrame(data=meanOf2019Data) 
df2020 = pd.DataFrame(data=meanOf2020Data) 
```

```python
def find_outliers(column):
    col_mean = df[column].mean()
    col_sd = np.std(df[column])
    outliers = []

    for item in df[column]:
        if item > (col_mean + (col_sd * 2)):
            outliers.append(item)
        elif item < (col_mean - (col_sd * 2)):
            outliers.append(item)

    return outliers

Watchtime_outliers = find_outliers('Watch time')

Streamtime_outliers = find_outliers('Stream time')

Followers_gained_percent_outliers = find_outliers('Followers gained percent')

Views_gained_outliers = find_outliers('Views gained')
print(Watchtime_outliers, Streamtime_outliers, Followers_gained_percent_outliers, Views_gained_outliers)
```

Through our outlier analysis we found a few outliers within our dataset. The outlier in the Watch time column, first outlier in the Followers gained percent, and second outlier in the Views gained column are all from month 15 which is the month after Covid started, and we believe it is important data to visualize and understand. Including row 15, we have 4 additional rows with outliers. Removing whole rows with outliers would only leave our data with 19 rows and would contradict and affect our time model, so we decided to keep them in our dataset.

```python
df7 = df[['Views gained', 'Followers gained percent']]
df7.columns = ['Views_gained', 'Followers_gained_percent']
(outcome, predictors) = patsy.dmatrices('Views_gained ~ Followers_gained_percent', df7) 
mod_log = sm.OLS(outcome, predictors)
res_log = mod_log.fit()
print(res_log.summary())
```

```python
df6 = df[['Watch time', 'Followers gained percent']]
df6.columns = ['Watch_time', 'Followers_gained_percent']
(outcome, predictors) = patsy.dmatrices('Watch_time ~ Followers_gained_percent', df6) 
mod_log = sm.OLS(outcome, predictors)
res_log = mod_log.fit()
print(res_log.summary())
```

We conducted an OLS regression for the combined dataframe of 2019 and 2020.Our results showed a p-value of 0.00 for both dataframes, meaning we can reject the null hypothesis. This means the predicator value “Followers gained percent” is related to changes in the response variable “Watch time” and “Views gained”.

```python
df8 = df[['Followers gained percent', 'Watch time', 'Stream time']]
df8.columns = ['Followers_gained_percent', 'Watch_time', 'Stream_time']
(outcome, predictors) = patsy.dmatrices('Followers_gained_percent ~ Watch_time + Stream_time', df8) 
mod_log = sm.OLS(outcome, predictors)
res_log = mod_log.fit()
print(res_log.summary())
```

In order to check if stream time is a potential confounding variable, we added the variable stream time to the equations of the OLS for the combined data frame for both watch time and viewers gained. It can still be seen that the p-values of both OLS for watch time and views gained was less than 0.05, meaning that stream time did not affect the significance. The stream time p-value for both was not significant, meaning that there was no relationship between stream time and the other variables.
